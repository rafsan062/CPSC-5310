# 2. Get MNIST Data

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X, y = mnist.data, mnist.target
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

# 3. Training SGDClassifier

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt

def plot_digit(image_data):
    image = image_data.reshape(28, 28)
    plt.imshow(image, cmap="binary")
    plt.axis("off")

some_digit = X[15]
plot_digit(some_digit)
plt.show()
```

```python
sgd_clf.predict([some_digit])
```

```python
sgd_clf.predict([X[12]])
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train, cv=3, scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(sgd_clf, X_train, y_train, cv=3)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_train, y_train_pred)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay.from_predictions(
    y_train, 
    y_train_pred, 
    ax=ax, 
    cmap="viridis", 
    normalize='true',  # This does the math for you!
    values_format='.1%' # This formats the text as percentages
)
plt.show()
```

# 4. Training Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)
```

```python
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train, cv=3,
                                    method="predict_proba")
y_probas_forest[14:17]
```

```python
cross_val_score(forest_clf, X_train, y_train, cv=3, scoring="accuracy")
```

```python
plt.figure(figsize=(9, 9))
for idx, image_data in enumerate(X[14:17]):
    plt.subplot(10, 10, idx + 1)
    plot_digit(image_data)
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
y_train_pred_forest = np.argmax(y_probas_forest, axis=1)
y_train_pred_forest[14:17]
```

```python
y_train = y_train.astype(np.uint8)
```

```python
cm_forest = confusion_matrix(y_train, y_train_pred_forest)
cm_forest
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay.from_predictions(
    y_train, 
    y_train_pred_forest, 
    ax=ax, 
    cmap="viridis", 
    normalize='true',  # This does the math for you!
    values_format='.1%' # This formats the text as percentages
)
plt.show()
```

# 5. Generating Augmented Data

```python
from scipy.ndimage import shift

def shift_image(image, dx, dy):
    image = image.reshape((28, 28))
    shifted_image = shift(image, [dy, dx], cval=0, mode="constant")
    return shifted_image.reshape([-1])
```

```python
import matplotlib.pyplot as plt

some_digit = X_train[1000]
shifted_digit = shift_image(some_digit, 5, 0) # Shift right by 5 pixels

plt.subplot(121)
plt.title("Original")
plt.imshow(some_digit.reshape(28, 28), cmap="binary")

plt.subplot(122)
plt.title("Shifted Right")
plt.imshow(shifted_digit.reshape(28, 28), cmap="binary")

plt.show()
```

```python
X_train_augmented = [image for image in X_train]
y_train_augmented = [label for label in y_train]

for image, label in zip(X_train, y_train):
    for dx, dy in ((1, 0), (-1, 0), (0, 1), (0, -1)):
        X_train_augmented.append(shift_image(image, dx, dy))
        y_train_augmented.append(label)

X_train_augmented = np.array(X_train_augmented)
y_train_augmented = np.array(y_train_augmented)

print(f"Original size: {len(X_train)}")
print(f"Augmented size: {len(X_train_augmented)}")
```

```python
shuffle_idx = np.random.permutation(len(X_train_augmented))

X_train_augmented = X_train_augmented[shuffle_idx]
y_train_augmented = y_train_augmented[shuffle_idx]
```

```python
forest_clf_ext = RandomForestClassifier(random_state=42)
```

```python
y_probas_forest_ext = cross_val_predict(forest_clf_ext, X_train_augmented, y_train_augmented, cv=3,
                                    method="predict_proba")
y_probas_forest[14:17]
```

```python
cross_val_score(forest_clf_ext, X_train_augmented, y_train_augmented, cv=3, scoring="accuracy")
```

```python
y_train_pred_forest_ext = np.argmax(y_probas_forest_ext, axis=1)
cm_forest_ext = confusion_matrix(y_train_augmented, y_train_pred_forest_ext)
cm_forest_ext
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay.from_predictions(
    y_train_augmented, 
    y_train_pred_forest_ext, 
    ax=ax, 
    cmap="viridis", 
    normalize='true',  # This does the math for you!
    values_format='.1%' # This formats the text as percentages
)
plt.show()
```

# 6. Accuracy Comparison

```python
from sklearn.metrics import accuracy_score
forest_clf.fit(X_train, y_train)
y_test_pred = forest_clf.predict(X_test)
y_test = y_test.astype(np.uint8)
final_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Final Test Accuracy - Random forest trained with Regular Dataset: {final_accuracy * 100:.2f}%")
```

```python
forest_clf_ext.fit(X_train_augmented, y_train_augmented)
y_test_pred_ext = forest_clf_ext.predict(X_test)
final_accuracy = accuracy_score(y_test, y_test_pred_ext)
print(f"Final Test Accuracy - Random forest trained with Augmented Dataset: {final_accuracy * 100:.2f}%")
```

# 7. Personal Reflections

From this notebook, I have learned that data augmentation, specifically image shifting, can significantly increase the training dataset size and often improve accuracy. The Random Forest Classifier consistently outperformed the SGDClassifier, demonstrating the benefit of more complex models for multiclass tasks. 

The notable increase in test accuracy from 96.91% to 98.04% after training with the augmented dataset clearly demonstrates that data augmentation (when applied effectively) allows the model to generalize better and achieve superior real world performance, even on unseen, nonaugmented test data. It highlights augmentation as a powerful technique to enhance model robustness and accuracy.
